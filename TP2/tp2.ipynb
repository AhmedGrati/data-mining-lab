```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('result.csv')
df.head(5)
```

# Drop Useless columns

```python
useless_cols = ['Date_time','Key_Run','Key_Fold','Key_Scheme_version_ID','Number_of_training_instances','Number_of_testing_instances',
'Number_correct','Number_incorrect','Number_unclassified','Percent_unclassified','True_positive_rate',
'Num_true_positives','False_positive_rate','Num_false_positives','True_negative_rate','Num_true_negatives','False_negative_rate','Num_false_negatives'
]
for col in useless_cols:
    df = df.drop(col, 1)
df.describe()
```

```python
knn = 'weka.classifiers.lazy.IBk'
df['Key_Scheme'] = df.apply(lambda x: 'Ibk'.__add__(x['Key_Scheme_options'].split(' ')[1]) if x['Key_Scheme'] == knn else x['Key_Scheme'], axis = 1)
df = df.drop('Key_Scheme_options',1)
df
```

# Group Results By Dataset

```python
df_groupby_datasets = df.groupby(['Key_Dataset','Key_Scheme']).mean()
df_groupby_datasets.index[0][1]
```

# Comparing Models Performance

```python
algos = []
for index in df_groupby_datasets.index:
    if(not index[1] in algos):
        algos.append(index[1])
print(algos)
```

```python
def compare_models(col: str, dataset: str):
    data = []
    for ind in df_groupby_datasets.index:
        if (ind[0] == dataset):
            data.append(df_groupby_datasets.loc[df_groupby_datasets.index == ind, df_groupby_datasets.columns == col].values[0])
    data = [item for sublist in data for item in sublist]
    fig = plt.figure(figsize = (20, 5))
    plt.bar(algos, data)
    plt.xlabel('Algorithm')
    plt.ylabel(col)
    plt.title(f'{dataset.capitalize()} - {col}')
```

```python
datasets = ['Glass','iris', 'weather', 'weather.symbolic']
important_cols = ['IR_precision', 'IR_recall', 'F_measure', 'Root_mean_squared_error', 'Elapsed_Time_training', 'Elapsed_Time_testing', 'Percent_correct']
for dataset in datasets:
    for col in important_cols:
        compare_models(col=col, dataset=dataset)
```

# Interpretations

## Glass Dataset Interpretation
The glass dataset contains 213 instances.<br>
Considering **Accuracy, Recall, Precision, and F-score**, the best algorithm for this dataset is **Random Forest.** However, This algorithm is the **slowest** one in both phases: Training and Testing, which seems logic because it's an ensemblist algorithm. In fact, it uses multiple decision trees in the training and testing phases.
